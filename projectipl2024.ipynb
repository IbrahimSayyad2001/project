Visualize the distribution of total tickets  across all matches


```python
sns.histplot(df['Total number of ticket sell'],kde=True)
plt.title('Distribution of Total Ticket sell')
plt.xlabel('Total Ticket sell')
plt.ylabel('Frequency')
plt.show()

```

 Create a bar plot showing the Corporate ticket price for each venue


```python
plt.figure(figsize=(15,6))
sns.barplot(x='Venue',y='Corporate Ticket Rate',data=df)
plt.title('Corporate ticket price for each venue')
```

 Plot the relationship between the number of adult tickets sold and the total revenue


```python
sns.scatterplot(x='Number of Adult Tickets',y='Total_number_of_tickets_sell',data=df)
plt.title('Relationship between Adult Tickets Sold and Total Revenue')
plt.xlabel('Number of Adult Tickets')
plt.ylabel('Total_number_of_tickets_sell ')
plt.show()
```

Create a count plot showing the number of matches scheduled on each day


```python
sns.countplot(x='Day',data=df,palette="rainbow")
plt.title('Number of matches schedule on each day')
plt.xlabel('Days')
plt.ylabel('count')
plt.show()
```

 Create a box plot of ticket prices for each category

```python
plt.figure(figsize=(11,8))
sns.boxplot(data=df[['Adult Ticket Rate','Senior Ticket Rate','Child Ticket Rate','VIP Ticket Rate','Corporate Ticket Rate','Student Ticket Rate']])
plt.title('Ticket prices for each Category')
plt.xlabel('Categories')
plt.xticks(rotation=45)
plt.ylabel('Ticket prices')
plt.show()
```

Plot the relationship between the match number and the total revenue

```python
plt.figure(figsize=(15,8))
sns.lineplot(x='Match No', y='Total_number_of_tickets_sell',data=df)
```

 Create a pie chart showing the distribution of matches across different venues


```python
matches_per_venue=df['Venue'].value_counts()
```

```python
plt.pie(matches_per_venue, labels=matches_per_venue.index, autopct='%1.1f%%')
plt.title('Distribution of matches Across Venue')
plt.show()
```
