What is the distribution of university scores in the dataset?



```python
plt.hist(df['score'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.title('Distribution of University Scores')
plt.show()
```

Bar plot of the top 10 countries with the most universities:



```python
universities_by_country = df['country'].value_counts().head(10)
universities_by_country
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=universities_by_country.index, y=universities_by_country.values, palette='rainbow')
plt.xlabel(' Country')
plt.ylabel('Number of Universities')
plt.title('Top 10 Countries with Most Universities')
plt.xticks(rotation=45)
plt.show()
```

How does the student-to-faculty ratio vary across different regions?


```python
plt.figure(figsize=(12, 8))
sns.countplot(x='region',data=df, hue='student_faculty_ratio' )
plt.title('Student-to-Faculty Ratio by Region')
plt.xlabel('Region')
plt.ylabel('Student-to-Faculty Ratio')
plt.xticks(rotation=45)
plt.show()
```

Conclusion: The box plot illustrates the variation in student-to-faculty ratios across different regions. Some regions have wider ranges of ratios, indicating more variability in student-to-faculty ratios within those regions.



What is the trend of scores over the years?


```python
plt.figure(figsize=(12, 8))
sns.lineplot(x='year', y='score', data=df)
plt.title('Trend of Scores Over the Years')
plt.xlabel('Year')
plt.ylabel('Score')
plt.show()
```

Conclusion: The line plot shows the trend of university scores over the years, which can indicate whether there have been any significant changes or trends in university performance
