```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
btc=pd.read_csv('/kaggle/input/bitcoin-dataset/bitcoin daatset_')
```

```python
btc
```

```python
btc.columns
```

```python
plt.figure(figsize=(10, 6))
plt.plot(btc.Month, btc['Closing Price'], marker='o', color='b', linestyle='-')
plt.title('Closing Price over Time')
plt.xlabel('Month')
plt.ylabel('Closing Price')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

***This line plot shows how the closing price varies over time (months). 
The x-axis represents the months, and the y-axis represents the closing price.
Markers are placed at each data point for better visualization****

```python
plt.figure(figsize=(10, 6))
plt.scatter(btc['Volume'], btc['Closing Price'], color='r', alpha=0.5)
plt.title('Volume vs. Closing Price')
plt.xlabel('Volume')
plt.ylabel('Closing Price')
plt.grid(True)
plt.show()
```

**This scatter plot shows the relationship between volume and closing price.
Each point represents a month's data. 
The x-axis represents volume, and the y-axis represents the closing price.
The alpha parameter controls the transparency of the points**

```python
plt.figure(figsize=(10, 6))
sns.histplot(btc['Change'], bins=20, kde=True, color='g')
plt.title('Distribution of Change')
plt.xlabel('Change')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

**This histogram shows the distribution of the 'Change' column.
The x-axis represents the change values, and the y-axis represents the frequency of occurrence.
The kde parameter adds a kernel density estimate to the plot.
**

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=btc['Market Cap'], color='m')
plt.title('Market Cap Distribution')
plt.xlabel('Market Cap')
plt.grid(True)
plt.show()
```

**This box plot shows the distribution of market capitalization.
The box represents the interquartile range (IQR), with the median marked by a line inside the box.
The whiskers extend to show the rest of the distribution.
Outliers are shown as individual points beyond the whiskers.
****

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=btc.Month, y=btc['Opening Price'], palette='viridis')
plt.title('Average Opening Price by Month')
plt.xlabel('Month')
plt.ylabel('Average Opening Price')
plt.grid(True)
plt.show()
```

***This bar plot shows the average opening price for each month.
The x-axis represents the months, and the y-axis represents the average opening price.
Each bar represents the average opening price for a specific month.*
****

```python
sns.barplot(x='Month', y='Volume', data=btc)
plt.title('Monthly Volume')
plt.xlabel('Month')
plt.ylabel('Volume')
plt.xticks(rotation=45)
plt.show()
```

Barplot is used to represent the distribution of a categorical variable.
In this plot, we are showing the monthly volume of transactions.
The x-axis represents the months, and the y-axis represents the volume.
Each bar represents the volume of transactions in a particular month.


```python
sns.scatterplot(x='Opening Price', y='Closing Price', data=btc)
plt.title('Opening vs Closing Price')
plt.xlabel('Opening Price')
plt.ylabel('Closing Price')
plt.show()
```

**Scatterplot is used to visualize the relationship between two continuous variables.
In this plot, we are showing the relationship between opening and closing prices.
Each point represents a month, with its corresponding opening and closing prices.
**

```python
sns.pairplot(btc[['Opening Price', 'Closing Price', 'Highest Price', 'Lowest Price']])
plt.suptitle('Pair Plot of Price Variables')
plt.show()
```

**Pair plot is used to visualize pairwise relationships between different variables in a dataset.
In this plot, we are showing the pairwise relationships between opening price, closing price, highest price, and lowest price.
Each scatterplot in the pair plot represents the relationship between two variables.
**

```python
plt.pie(btc['Market Cap'], labels=btc['Month'], autopct='%1.1f%%')
plt.title('Market Cap Distribution')
plt.show()
```

Pie chart is used to represent the proportion of each category in a dataset.
In this plot, we are showing the distribution of market capitalization across different months.
Each slice of the pie represents the proportion of market capitalization for a particular month.
